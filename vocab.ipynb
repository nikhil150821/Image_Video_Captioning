```python
# Cell 2: Import Libraries
import os
import pickle
from collections import Counter
```

```python
# Cell 3: Load Caption Pairs
with open('flicker/image_caption_pairs.pkl', 'rb') as f:
    image_pairs = pickle.load(f)

with open('msrvtt_data/frame_caption_pairs.pkl', 'rb') as f:
    video_pairs = pickle.load(f)
```

```python
# Cell 4: Combine and Tokenize Captions
all_captions = [caption.lower().strip() for _, caption in image_pairs + video_pairs]
tokens = [word for cap in all_captions for word in cap.split()]
```

```python
# Cell 5: Build Vocabulary
counter = Counter(tokens)
threshold = 1
words = [word for word, cnt in counter.items() if cnt >= threshold]

vocab = {
    '<pad>': 0,
    '<start>': 1,
    '<end>': 2,
    '<unk>': 3
}
for i, word in enumerate(words, start=4):
    vocab[word] = i

# Save Vocabulary
with open('vocab.pkl', 'wb') as f:
    pickle.dump(vocab, f)

print("✅ Vocabulary created with", len(vocab), "words and saved to vocab.pkl")
```
